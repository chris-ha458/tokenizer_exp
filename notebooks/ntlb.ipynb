```python
import tokenizers
```

```python
tokenizer_path = "/home/karyo/corpus/models/06_02/tokenizer.json"
```

```python
tokenizer = tokenizers.Tokenizer.from_file(tokenizer_path)
```

```python
vocab_list = list(tokenizer.get_vocab())
```

```python
encoded_list = tokenizer.encode_batch(vocab_list)
```

```python
id_list = list(map(lambda token:token.ids,encoded_list))
```

```python
tokenizer.decode_batch(id_list)
```
