```python
json_path = "/home/karyo/corpus/data/mc4_310_ko/"
```

```python
os.listdir(json_path)
```

```python
def process_file(file):
    if not file.endswith(".json"): return
    output_name = os.path.splitext(file)[0] + ".jsonl"
    output_path = os.path.join(json_path,output_name)
    input_path = os.path.join(json_path, file)
    with jsonlines.open(input_path,mode="r") as reader, jsonlines.open(output_path,mode="w") as writer:
        #print(f"Processing {file}\n")
        for line in reader:
            writer.write({"text": line['text']})
```

```python
files = os.listdir(json_path)
with Pool() as pool:
    list(tqdm(pool.imap(process_file, files), total=len(files)))
```
